```python
# get the required libraries 
import numpy as np 
import pandas as pd 


import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# get the data 
df = pd.read_csv('data/Car_Insurance_Claim.csv')
```

```python
# convert the pandas into numpy
df.info()
```

```python
# select the appropriate column to put on the correlation matrix , make a list 
```

```python
# this is a lot to unpack
# let's just take a number for correlation.

corr = df.corr()
corr
```

```python
corr_matrix = df.corr().iloc[1:, 1:]
fig, ax = plt.subplots(figsize=(22, 10))
ax = sns.heatmap(corr_matrix,annot=True,linewidths=0.5,fmt=".2f",cmap="YlGn");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5);
```

```python
#df = df.dropna()
x = df.select_dtypes(include=['float64' , 'int64'])
x
y_tr = df['OUTCOME'].values
x_tr = x.drop('OUTCOME', axis =1).values
x_tr
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
logr = DecisionTreeClassifier(max_depth=3)
logr = logr.fit(x_tr, y_tr)
print(sum(logr.predict(x_tr) == y_tr)/len(y_tr))
```
